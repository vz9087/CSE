### 1.a. Write a python program to find the best of two test average marks out of three test’s marks accepted from the user.

```python
test1 = float (input ("Enter Test 1 marks:"))     # Test 1 marks
test2 = float (input ("Enter Test 2 marks:"))     # Test 2 marks
test3 = float (input ("Enter Test 3 marks:"))     # Test 3 marks

if test1 <= test2 and test1 <= test3 :     # Compare for Test 1
 min_score = test1
elif test2 <= test1 and test2 <= test3 :   # Compare for Test 2
 min_score = test2
else :                                     # Final compare for Test 3
 min_score = test3
    
average = (test1 + test2 + test3 - min_score) / 2  # Average case

print ("\nThe Average of Best of Two Tests is", average) # Print average of best of two tests

# Use 3 inputs of any amount of digits to get output. Ex: 45, 20, 40 = 42.5
```

### 2.a. Defined as a function F as Fn = Fn-1 + Fn-2, write a Python program which accepts a value for N (where N >0) as input and pass this value to the function. Display suitable error message if the condition for input value is not followed.

```python
def Fibonacci (n) :
    if n <= 0   :    print ("Incorrect input!")    # First Fibonacci number is 0
    elif n == 1 :    return 0                      # Second Fibonacci number is 1
    elif n == 2 :    return 1                      # Third Fibonacci number is 2
    else        :    return Fibonacci(n-1)+Fibonacci(n-2)
        
n = int (input ("Enter a number:"))    # Give an input for Fibonacci

print ("Fibonacci number is", Fibonacci (n))    # Print the Fibonacci number

# Use numbers like 1, 2, 3, 4, 10 = Fibonacci number is 0, 1, 1, 2, 34
# Warning: Using higher positive inputs require more time for calculation and may break the kernel!
# Incorrect usage of inputs: -2, -100, -999, -3092 = Incorrect input!
```

### 3.a. Write a Python program that accepts a sentence and find the number of words, digits, uppercase letters and lowercase letters.

```python
s = input ("Enter a sentence:")

w, d, u, l = 0, 0, 0, 0

l_w = s.split ()
w = len (l_w)

for c in s : 
    if c.isdigit   () :   d = d + 1
    elif c.isupper () :   u = u + 1
    elif c.islower () :   l = l + 1
                 
print ("\nNo. of Words found are", w)
print ("No. of Digits are", d)
print ("No. of Uppercase letters are", u)
print ("No. of Lowercase letters are", l)

# Enter a mix of words, digits, upper and lower case letters.
# Special characters are also accepted as words.

# Simple use case:
# Enter a sentence: Python is easy in 2025!

# No. of Words found are 5
# No. of Digits are 4
# No. of Uppercase letters are 1
# No. of Lowercase letters are 13
```

### 3.b. Write a Python program to find the string similarity between two given strings.

```python
import difflib      # import the library that verif(*y)ies the difference

string_1 , string_2 = "Welcome to Computer Science" , "Welcome to Computer Network"  # pre-defined strings

sequence = difflib.SequenceMatcher (a = string_1.lower (), b = string_2.lower ())    # Verify mactching cases

difference = sequence.ratio () * 100        # Calculate difference ration
difference = round (difference, 1)          # Round the value to one decimal digit from he right side

print ("Similarity between two said strings is " + str (difference))   # Print the difference
```

### 7. Write a Python program which explains uses of customizing seaborn plots with Aesthetic functions.

```python
# Import necessary libraries: Matplotlib, Numpy, Seaborn

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def sinplot (n = 10) :                # Define the sinplot function
    x = np.linspace( 0, 14, 100)           
    
    for i in range (1, n + 1) :       # Set the plot range and plot the sine wave
        plt.plot (x, np.sin (x + i * .5) * (n + 2 - i))

sns.set_theme ()          # Set the theme

sns.set_context ("notebook", font_scale = 1.5, rc = {"lines.linewidth" : 2.5})    # Setup context variables

sinplot ()                # Call the sinplot () function

plt.title ('Seaborn Plot with Aesthetic Functions', size = 20)                    # Add a title to display

plt.show ()               # Show the seaborn plot
```

### 8. Write a Python program to explain working with Bokeh Line Graph using Annotations and Legends.

```python
# Import necessary libraries

import numpy as np
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show
from bokeh.models import Label

# Generate data

x = np.linspace (0, 4 * np.pi, 100)
y = np.sin (x)

# Define interactive tools

TOOLS = "pan, wheel_zoom, box_zoom, reset, save, box_select"

# First plot: using scatter markers and legend

p1 = figure (title = "Example 1: Markers", tools = TOOLS)

p1.scatter (x, y, marker = "circle", size = 6, legend_label = "sin (x)", color = "blue")
p1.scatter (x, 2 * y, marker = "circle", size = 6, legend_label = "2 * sin (x)", color = "orange")
p1.scatter (x, 3 * y, marker = "circle", size = 6, legend_label = "3 * sin (x)", color = "green")

p1.legend.title = "Markers"
p1.legend.location = "top_right"

# Add annotation to first plot

label1 = Label (x = 1, y = 3 * np.sin (1), text = "Peak of 3 * sin (x)", text_color = "green")
p1.add_layout (label1)

# Second plot: using lines and scatter with square markers

p2 = figure (title = "Example 2: Lines & Shapes", tools = TOOLS)

p2.scatter (x, y, marker = "circle", size = 4, legend_label = "sin (x)", color = "blue")
p2.line (x, y, legend_label = "sin (x)", line_width = 2, line_color = "blue")

p2.line (x, 2 * y, legend_label = "2 * sin (x)", line_dash = "dashed", line_color = "orange", line_width = 2)

p2.scatter (x, 3 * y, marker = "square", size = 6, legend_label = "3 * sin (x)", fill_color = None, line_color = "green")
p2.line (x, 3 * y, legend_label = "3 * sin (x)", line_color = "green", line_width = 2)

p2.legend.title = "Lines"
p2.legend.location = "top_right"

# Add annotation to second plot

label2 = Label (x = 3, y = np.sin (3), text = "sin (x) at x = 3", text_color = "blue")

p2.add_layout (label2)

# Display both plots side by side

show (gridplot ([p1, p2], ncols = 2, width = 400, height = 400))

# Run = Shift + Enter 

# Output may take you to a new tab if used with "Run All cells"
```

### 9. Write a Python program to draw 3D Plots using Plotly Libraries.

```python
# Import PLotly and Numpy libraries

import plotly.graph_objects as go
import numpy as np

x = np.linspace (-5, 5, 100)
y = np.linspace (-5, 5, 100)

x, y = np.meshgrid (x, y)

z = np.sin (np.sqrt (x ** 2 + y ** 2))

fig = go.Figure (data = [go.Surface (z = z, x = x, y = y)])

fig.update_layout (     scene = dict (xaxis_title = 'X Axis', yaxis_title = 'Y Axis', zaxis_title = 'Z Axis'),
                        margin = dict (l = 0, r = 0, b = 0, t = 40),
                        title = '3D Surface Plot of sin (sqrt (x ^ 2 + y ^ 2)) is shown below'    )

fig.show ()
```

### 10.b. Write a Python program for creating Maps using Plotly Libraries. 

```python
# Importing plotly.express library

import plotly.express as px

df      = px.data.election ()              # Access the election data
geojson = px.data.election_geojson ()      # Access and enable the geojson view

# Declare choropleth_map to direct the map view

fig = px.choropleth_map (df, geojson = geojson, color = "Bergeron",
                            locations = "district", featureidkey = "properties.district",
                            center = {"lat" : 45.5517, "lon" : -73.7073},
                            map_style = "carto-positron", zoom = 9)

fig.update_layout (margin = {"r" : 0, "t" : 0, "l" : 0, "b" : 0})      # Justify the map view

fig.show ()          # Display the map
```
